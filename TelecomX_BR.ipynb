VerificaÃ§Ã£o de Erros de FormataÃ§Ã£o

```python
print("\n### Tipos de Dados ###")
print(df.dtypes)

# Verificar formataÃ§Ã£o em colunas
for col in df.columns:
    # Aplicando verificaÃ§Ãµes de string somente a colunas de tipo de objeto que nÃ£o sejam dicionÃ¡rios aninhados
    if df[col].dtype == 'object' and col not in ['customer', 'phone', 'internet', 'account']:
        unique_values = df[col].unique()
        print(f"\nValores Ãºnicos em '{col}': {unique_values[:10]}")  # Mostrar atÃ© 10 valores
        if df[col].astype(str).str.contains(r'^\s+|\s+$', regex=True, na=False).any():
            print(f"A coluna '{col}' contÃ©m espaÃ§os em branco no inÃ­cio ou fim.")
        if df[col].astype(str).str.contains(r'[^a-zA-Z0-9\s]', regex=True, na=False).any():
            print(f"A coluna '{col}' contÃ©m caracteres especiais.")

    elif df[col].dtype in ['int64', 'float64']:  # Colunas numÃ©ricas
        if (df[col] < 0).any():
            print(f"A coluna '{col}' contÃ©m valores negativos, verifique se isso Ã© esperado.")
        print(f"EstatÃ­sticas de '{col}':")
        print(df[col].describe())
```

```python
print("\n### Resumo dos Problemas Encontrados ###")
problems = []
if missing_values.sum() > 0:
    problems.append("Valores ausentes detectados em algumas colunas.")
if duplicates > 0:
    problems.append("Linhas duplicadas encontradas.")
for col in df.columns:
    # Excluir colunas de dicionÃ¡rio de verificaÃ§Ãµes de string e capitalizaÃ§Ã£o
    if col not in ['customer', 'phone', 'internet', 'account']:
        if df[col].dtype == 'object':
            if df[col].astype(str).str.contains(r'^\s+|\s+$', regex=True, na=False).any():
                problems.append(f"EspaÃ§os em branco detectados na coluna '{col}'.")
            if df[col].astype(str).str.contains(r'[^a-zA-Z0-9\s]', regex=True, na=False).any():
                 problems.append(f"A coluna '{col}' contÃ©m caracteres especiais.")
            if df[col].astype(str).str.lower().nunique() < df[col].nunique():
                problems.append(f"InconsistÃªncias de capitalizaÃ§Ã£o na coluna '{col}'.")

        elif df[col].dtype in ['int64', 'float64']:  # Colunas numÃ©ricas
            if (df[col] < 0).any():
                problems.append(f"Valores negativos detectados na coluna '{col}', verifique se sÃ£o vÃ¡lidos.")


if problems:
    print("Problemas identificados:")
    for problem in problems:
        print(f"- {problem}")
else:
    print("Nenhum problema crÃ­tico identificado.")
```

 CorreÃ§Ã£o de Valores Ausentes

```python
import pandas as pd
import requests
import json
from datetime import datetime
```

```python
print("\n### CorreÃ§Ã£o de Valores Ausentes ###")

# Identificar valores ausentes
missing_values = df.isnull().sum()
print("Valores ausentes antes da correÃ§Ã£o:")
print(missing_values[missing_values > 0])
```

```python
# ImputaÃ§Ã£o
for col in df.columns:
    if df[col].isnull().sum() > 0:  # Apenas colunas com valores ausentes
        if df[col].dtype in ['int64', 'float64']:  # Colunas numÃ©ricas
            median_value = df[col].median()
            df[col].fillna(median_value, inplace=True)
            print(f"Coluna '{col}' preenchida com a mediana: {median_value}")
        elif df[col].dtype == 'object':  # Colunas categÃ³ricas
            mode_value = df[col].mode()[0]
            df[col].fillna(mode_value, inplace=True)
            print(f"Coluna '{col}' preenchida com a moda: {mode_value}")
        elif 'date' in col.lower() or df[col].dtype == 'datetime64[ns]':  # Colunas de data
            default_date = datetime.now().strftime('%Y-%m-%d')
            df[col].fillna(default_date, inplace=True)
            print(f"Coluna '{col}' preenchida com a data padrÃ£o: {default_date}")
```

```python
# Verificar valores ausentes apÃ³s correÃ§Ã£o
print("\nValores ausentes apÃ³s correÃ§Ã£o:")
print(df.isnull().sum())
```

CorreÃ§Ã£o de InconsistÃªncias CategÃ³ricas

```python
print("\n### CorreÃ§Ã£o de InconsistÃªncias CategÃ³ricas ###")

# DicionÃ¡rio de valores esperados (ajuste conforme dicionÃ¡rio de dados, se disponÃ­vel)
expected_values = {
    'plano': ['prÃ©-pago', 'pÃ³s-pago', 'controle'],
    'status': ['ativo', 'cancelado', 'suspenso']
}

categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    if col in expected_values:
        # Verificar valores fora do esperado
        invalid_values = df[~df[col].isin(expected_values[col])][col].unique()
        if len(invalid_values) > 0:
            print(f"Valores invÃ¡lidos em '{col}': {invalid_values}")
            # Substituir valores invÃ¡lidos pela moda
            mode_value = df[col].mode()[0]
            df.loc[~df[col].isin(expected_values[col]), col] = mode_value
            print(f"Valores invÃ¡lidos em '{col}' substituÃ­dos pela moda: {mode_value}")
```

```python
# Verificar distribuiÃ§Ã£o apÃ³s correÃ§Ã£o
print(f"\nDistribuiÃ§Ã£o de valores em '{col}' apÃ³s correÃ§Ã£o:")
print(df[col].value_counts(dropna=False))
```

```python
print("\n### CriaÃ§Ã£o da Coluna Contas_Diarias ###")

# Identificar a coluna de faturamento mensal (ajuste conforme o nome real)
monthly_column = None
# Pesquise a coluna de cobranÃ§as mensais no DataFrame normalizado
for col in df_normalizado.columns:
    if 'monthly' in col.lower():
        monthly_column = col
        break

if monthly_column:
    # Calcular Contas_Diarias (valor mensal / 30)
    # Garantir que as divisÃµes mensais da coluna sejam numÃ©ricas antes
    df_normalizado[monthly_column] = pd.to_numeric(df_normalizado[monthly_column], errors='coerce')

    # Removendo as linhas onde Charges.Monthly Ã© NaN apÃ³s a coerÃ§Ã£o para evitar erros na divisÃ£o
    df_normalizado.dropna(subset=[monthly_column], inplace=True)
    df_normalizado['Contas_Diarias'] = df_normalizado[monthly_column] / 30
    print(f"Coluna 'Contas_Diarias' criada com base em '{monthly_column}' (dividido por 30).")
else:
    raise ValueError("Coluna de faturamento mensal nÃ£o encontrada no DataFrame normalizado. Verifique os nomes das colunas e ajuste o cÃ³digo.")

# Visualizar as primeiras linhas com a nova coluna
print("\nPrimeiras 5 linhas com a nova coluna 'Contas_Diarias':")
# Display the relevant columns from the normalized DataFrame
display(df_normalizado[[monthly_column, 'Contas_Diarias']].head())

# Verificar estatÃ­sticas bÃ¡sicas da nova coluna
print("\nEstatÃ­sticas da coluna 'Contas_Diarias':")
print(df_normalizado['Contas_Diarias'].describe())
```

ValidaÃ§Ã£o da Nova Coluna

```python
print("\n### ValidaÃ§Ã£o da Coluna Contas_Diarias ###")

# Verificar valores nulos
null_values = df_normalizado['Contas_Diarias'].isnull().sum()
print(f"Valores nulos em 'Contas_Diarias': {null_values}")
```

```python
# Verificar valores negativos
negative_values = (df_normalizado['Contas_Diarias'] < 0).sum()
if negative_values > 0:
    print(f"AtenÃ§Ã£o: {negative_values} valores negativos encontrados em 'Contas_Diarias'.")
    # Corrigir valores negativos (assumindo que sÃ£o erros)
    df_normalizado.loc[df_normalizado['Contas_Diarias'] < 0, 'Contas_Diarias'] = df_normalizado['Contas_Diarias'].abs()
    print("Valores negativos corrigidos (convertidos para absolutos).")
```

```python
# Verificar valores extremos (ex.: muito altos)
q1 = df_normalizado['Contas_Diarias'].quantile(0.25)
q3 = df_normalizado['Contas_Diarias'].quantile(0.75)
iqr = q3 - q1
outliers = df_normalizado[(df_normalizado['Contas_Diarias'] > q3 + 1.5 * iqr) | (df_normalizado['Contas_Diarias'] < q1 - 1.5 * iqr)]
if not outliers.empty:
    print(f"\nPossÃ­veis outliers em 'Contas_Diarias': {len(outliers)} valores.")
    print(outliers[['Contas_Diarias']].head())
```

```python
# Resumo final
print("\nResumo apÃ³s validaÃ§Ã£o:")
print(df_normalizado['Contas_Diarias'].describe())
```

```python
service_columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies'
]

# Create a copy to avoid SettingWithCopyWarning if df_normalizado is a view
df_normalizado = df_normalizado.copy()

# Initialize 'TotalServices' column with zeros
df_normalizado['TotalServices'] = 0

# Count 'Yes' values for each service column
for col in service_columns:
    # Ensure the column exists in df_normalizado before processing
    if col in df_normalizado.columns:
        # For PhoneService, count 'Yes'. 'No' and 'No phone service' are not having the service.
        if col == 'PhoneService':
            df_normalizado['TotalServices'] += (df_normalizado[col] == 'Yes').astype(int)
        # For InternetService, count 'DSL' and 'Fiber optic'. 'No' and 'No internet service' are not having the service.
        elif col == 'InternetService':
             df_normalizado['TotalServices'] += df_normalizado[col].apply(lambda x: 1 if x in ['DSL', 'Fiber optic'] else 0)
        # For other services, count 'Yes'. 'No' and 'No internet service' are not having the service.
        else:
            df_normalizado['TotalServices'] += (df_normalizado[col] == 'Yes').astype(int)
    else:
        print(f"Warning: Service column '{col}' not found in df_normalizado.")


# Display the first few rows with the new column
display(df_normalizado[['PhoneService', 'MultipleLines', 'InternetService',
                        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                        'TechSupport', 'StreamingTV', 'StreamingMovies',
                        'TotalServices']].head())
```
